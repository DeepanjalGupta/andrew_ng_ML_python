```python
%matplotlib inline
import numpy as np
import pandas as pd
import random

import matplotlib.pyplot as plt
import scipy.io #Used to load the OCTAVE *.mat files
from scipy.special import expit
from scipy import optimize
from sklearn import preprocessing
from sklearn import svm
```

```python
data1 = scipy.io.loadmat('/Users/deepanjal.gupta/Documents/Learning/Stanford Machine Learning- Andrew Ng/Week 7/machine-learning-ex6/ex6/ex6data1.mat')

X = data1['X']
y = data1['y']

#splitting the data into +vs and negative examples
x1_positive = X[:,0:1][y ==1]
x2_positive = X[:,1:2][y ==1]

x1_negative = X[:,0:1][y ==0]
x2_negative = X[:,1:2][y ==0]
```

```python
plt.figure(figsize = (5,5))
plt.grid(True)
plt.scatter(x1_positive,x2_positive,marker = '+',c = 'g',label = 'Admitted')
plt.scatter(x1_negative,x2_negative,marker = '+',c = 'r',label = 'Not Admitted')

plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc = 3)
plt.show()
```

## 1. Linear Kernel 

```python
#overfit fit
linear_svm = svm.SVC(C=100, kernel= 'linear')
linear_svm.fit(X,y.flatten())
x1_min, x1_max = min(X[:,0]) - 1, max(X[:,0]) + 1
x2_min, x2_max = min(X[:,1]) - 1, max(X[:,1]) + 1
xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, 0.02),
                         np.arange(x2_min, x2_max, 0.02))

y_plot = linear_svm.predict(np.concatenate((xx1.flatten().reshape(-1,1),xx2.flatten().reshape(-1,1)),axis = 1)).reshape(xx1.shape)


plt.figure(figsize = (5,5))
mycontour = plt.contour( xx1, xx2, y_plot, [0])

plt.grid(True)
plt.scatter(x1_positive,x2_positive,marker = '+',c = 'g',label = 'Admitted')
plt.scatter(x1_negative,x2_negative,marker = '+',c = 'r',label = 'Not Admitted')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('C=1')
plt.legend(loc = 3)
plt.show()
```
